```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
print(df.shape)
```

```python
df.info()
```

```python
df = df.dropna()
print(df.shape)
```

```python
df.head()
```

```python
df1 = df[['popularity'] + ['budget']].groupby('budget').max()
df1
```

```python
df1.plot(kind='area')
#Популярность и бюджет. Бюджет, в целом, напрямую не влияет на популярность.
```

```python
df2 = df[['vote_average'] + ['budget']].groupby('budget').max()
df2
```

```python
ax = df['vote_average'].hist()
ax.set_title('Votes distribution')
ax.set_xlabel('Votes')
ax.set_ylabel('Count')
#Люди чаще всего ставят оценку от 5 до 7
```
